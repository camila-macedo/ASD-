```python
import os
import pandas as pd
from scipy.signal import butter, filtfilt

# Diretórios de leitura e escrita (exemplo, ajuste para o seu caso)
d_r = '/Users/camilamacedo/Desktop/MBA/TCC/paper-autism/Matriz_Corr/HC'
d_w = '/Users/camilamacedo/Desktop/MBA/TCC/paper-autism/Matriz_Corr/spearman-hc'

# Número total de arquivos a serem lidos (ASD)
n_files = 257

for i in range(n_files + 1):
    # Cria o nome do arquivo dinamicamente com base no valor de i
    file_r = f'HC{i}.csv'
    path_r = os.path.join(d_r, file_r)

    # Realiza a leitura do arquivo
    dataframe = pd.read_csv(path_r, header=None)

    # Realiza o cálculo da matriz de correlação (Spearman)
    correlation_matrix = dataframe.corr(method='spearman')

    # Exporta a matriz de correlação para um arquivo CSV
    file_w = f'MatrizCorrelacao_{file_r}'
    path_w = os.path.join(d_w, file_w)
    correlation_matrix.to_csv(path_w, header=False, index=False)

    print(f'Calculada e exportada a matriz de correlação para {file_r}')
```
